```python
# importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['figure.figsize'] = [10, 10] # setting figure size
```

```python
df = pd.read_csv('./Visual_text.csv')
df['num_parameters'] = list(map(int,df['num_parameters']))

df['train_loss'] = list(map(float,df['train_loss']))
df['train_loss'] = [np.log10(x) for x in df['train_loss']]

df['val1_loss'] = list(map(float,df['val1_loss']))
df['val1_loss'] = [np.log10(x) for x in df['val1_loss']]
```

```python
x = df.time
y = df.train_loss
y_val = df.val1_loss
area = (df.num_parameters)*300

sns.scatterplot(x, y, hue=df.model, legend=False, size=area, sizes=(500, 2000), alpha=.3)
sns.scatterplot(x, y_val, hue=df.model, alpha=1)

for i in range(len(df)):
    plt.text(df.iloc[i]['time'], df.iloc[i]['train_loss'], df.iloc[i].model)
```
